# Global Population Trend Analysis [2016-2022]

**Problem Statement**

Analyze the Data of Global Population to undersatnd the trend.

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\STUDY\Data Science\All Data Science Projects\Global Population Trend Analysis\Global Population Trends(2016-2022).csv")
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.replace('-',np.nan)
```

```python
df.head()
```

```python
df = df.replace(',','',regex=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Total Population'] = df['Total Population'].astype(float)
df['Urban Population'] = df['Urban Population'].astype(float)
df['Rural Population'] = df['Rural Population'].astype(float)
df['Population Density'] = df['Population Density'].astype(float)
df['Life Expectancy'] = df['Life Expectancy'].astype(float)
df['Fertility Rate'] = df['Fertility Rate'].astype(float)
df['Infant Mortality Rate'] = df['Infant Mortality Rate'].astype(float)
df['Growth Rate'] = df['Growth Rate'].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Total Population'].fillna(df['Total Population'].mean() , inplace = True)
df['Urban Population'].fillna(df['Urban Population'].mean() , inplace = True)
df['Rural Population'].fillna(df['Rural Population'].mean() , inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Total Population'] = df['Total Population'] / 1e6
```

```python
df.head()
```

```python
df.head()
```

```python
df['Urban Population'] = df['Urban Population'] / 1e6
df['Rural Population'] = df['Rural Population'] / 1e6
```

```python
df.head()
```

```python
df.rename(columns={'Total Population': 'Total Population( In Millions)'}, inplace=True)
df.rename(columns={'Urban Population': 'Urban Population( In Millions)'}, inplace=True)
df.rename(columns={'Rural Population': 'Rural Population( In Millions)'}, inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 10))
plt.hist(df['Year'], bins=5, weights=df['Total Population( In Millions)'], color='blue', edgecolor='black')
plt.xlabel('Year')
plt.ylabel('Population in Millions')
plt.title('Year-wise Population Trend')
plt.show()
```

**Q5. Are there regions with high population density but low urbanization?**

```python
high_density_threshold = df['Population Density'].max()
low_urbanization_threshold = df['Urban Population( In Millions)'].min()
```

```python
filtered_regions = df[(df['Population Density'] > high_density_threshold) & (df['Urban Population( In Millions)'] < low_urbanization_threshold)]
```

```python
print('Regions with high populatoin density but low urbanization are : ')
print(filtered_regions[['Country','Population Density','Urban Population( In Millions)']])
```

Here in this dataset there are no countries with high population density and low urbanization

**Q7. How does life expectancy correlate with urbanization and access to healthcare?**

```python
# correlation between burban pop and net life expectancy
correlation = df['Urban Population( In Millions)'].corr(df['Life Expectancy'])

print(f"Pearson's Correlation Coefficient: {correlation}")
```

```python
# correlation between burban pop and net life expectancy
correlation = df['Rural Population( In Millions)'].corr(df['Life Expectancy'])

print(f"Pearson's Correlation Coefficient: {correlation}")
```

**Q10. How does the fertility rate vary by country and region?**

**Q12. What is the global population growth rate, and how has it fluctuated over time?**
